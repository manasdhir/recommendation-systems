```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
pd.set_option('display.max_columns', 100)
df = pd.read_csv('https://query.data.world/s/uikepcpffyo2nhig52xxeevdialfl7')
df.head()
```

```python
df = df[['Title','Genre','Director','Actors','Plot']]
df.head()
```

```python
df['Actors'] = df['Actors'].map(lambda x: x.split(',')[:3])
```

```python
df['Actors']
```

```python
df['Genre'] = df['Genre'].map(lambda x: x.lower().split(','))#as python can consider crime ,CRIME and Crime as different genres
```

```python
df['Director'] = df['Director'].map(lambda x: x.split(' '))
df['Director'].head()
```

```python
for index, row in df.iterrows():
    row['Actors'] = [x.lower().replace(' ','') for x in row['Actors']]
    row['Director'] = ''.join(row['Director']).lower()
```

```python
keywords_list=[]
for index, row in df.iterrows():
    plot = row['Plot']
    r = Rake()
    r.extract_keywords_from_text(plot)
    key_words_dict_scores = r.get_word_degrees()
    keywords = list(key_words_dict_scores.keys())
    keywords_list.append(keywords)
df['Key_words'] = keywords_list
```

```python
df.drop(columns = ['Plot'], inplace = True)
```

```python
df.set_index('Title', inplace = True)
df.head()
```

```python
df['bag_of_words'] = ''
columns = df.columns
for index, row in df.iterrows():
    words = ''
    for col in columns:
        if col != 'Director':
            words = words + ' '.join(row[col])+ ' '
        else:
            words = words + row[col]+ ' '
    row['bag_of_words'] = words

df.drop(columns = [col for col in df.columns if col!= 'bag_of_words'], inplace = True)
```

this code adds all the values in a particular column with spaces in between them, a separate method is used for the director column as it only has 1 value thus if the same method is used for it then all the letters of the director name will be split with a space in between

```python
df.head()
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df['bag_of_words'])
```

```python
indices = pd.Series(df.index)
indices.head()
```

Count vectorization makes columns of each unique word present in all of the data and then creates vectors of the same length for all the different entries.The matrix obtained is known as document term matrix.

so our objective was to recommend x,y,z movies to a person who as watched a particular movie so we will find the cosine similarity between all components of the matrix itself

```python
cosine_sim = cosine_similarity(count_matrix, count_matrix)
cosine_sim
```

cosine similarity 1 for all the diagonals as at the diagonals the movie is being compared with itself

```python
def recommendations(title, cosine_sim = cosine_sim):

    recommended_movies = []

    # gettin the index of the movie that matches the title
    idx = indices[indices == title].index[0]

    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    # getting the indexes of the 10 most similar movies
    top_10_indexes = list(score_series.iloc[1:11].index)
    print(top_10_indexes)

    # populating the list with the titles of the best 10 matching movies
    for i in top_10_indexes:
        recommended_movies.append(list(df.index)[i])

    return recommended_movies
```

```python
recommendations("Fargo")
```

```python
recommendations("The Godfather: Part II")
```
